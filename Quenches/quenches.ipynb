```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Define pressure values
pressures = np.arange(0, 11000, 1000)

# Create subplots (2 rows, ceil(pressures/2) columns)
fig, ax = plt.subplots(2, int((len(pressures) + 1) / 2), sharex=True, sharey=True)
plt.subplots_adjust(wspace=0.2, hspace=0.3)

# Define colors
# Define colors
new_color = 'seagreen'
colors = [
    (0.2980392156862745, 0.4470588235294118, 0.6901960784313725),
    (0.8666666666666667, 0.5176470588235295, 0.3215686274509804),
    (0.3333333333333333, 0.6588235294117647, 0.40784313725490196),
    (0.7686274509803922, 0.3058823529411765, 0.3215686274509804),
    (0.5058823529411764, 0.4470588235294118, 0.7019607843137254),
    (0.5764705882352941, 0.47058823529411764, 0.3764705882352941),
    (0.8549019607843137, 0.5450980392156862, 0.7647058823529411),
    (0.5490196078431373, 0.5490196078431373, 0.5490196078431373),
    (0.8, 0.7254901960784313, 0.4549019607843137),
    (0.39215686274509803, 0.7098039215686275, 0.803921568627451)
] + [new_color]

# Loop through pressures and plot histograms
for j, pressure in enumerate(pressures):
    filename = f"QuenchT-230K/{pressure}atm_28/thermo.txt"
    
    try:
        data = np.genfromtxt(filename)
        phi = np.concatenate((data[::100, 3],-data[::100, 3]))  # Extract relevant column
        
        l = int(2 * j / len(pressures))  # Row index
        m = j % int((len(pressures) + 1) / 2)  # Column index
        
        # Plot histogram
        ax[l, m].hist(phi, color=colors[j], label=f"{pressure} atm", density=True)
        ax[l, m].text(0, 0.36, '{0:3.1f}'.format(pressure*0.000101325) + " GPa", color=colors[j], ha='center')
        
        if m == 0:
            ax[l, m].set_ylabel("Probability Density")
        if l == 1:
            ax[l, m].set_xlabel("$\\phi$")

    except Exception as e:
        print(f"Error reading {filename}: {e}")

# Adjust last subplot if needed
ax[-1, -1].remove()
ax[0, -1].set_xlabel("$\\phi$")

plt.ylim([0, 0.35])
plt.savefig("figs/PD-230K.pdf", bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Define pressure values
pressures = np.array([0, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000])

# Create subplots (2 rows, ceil(pressures/2) columns)
fig, ax = plt.subplots(2, int((len(pressures) + 1) / 2), sharex=True, sharey=True)

plt.subplots_adjust(wspace=0.2, hspace=0.3)

# Define colors
new_color = 'seagreen'
colors = [
    (0.2980392156862745, 0.4470588235294118, 0.6901960784313725),
    (0.8666666666666667, 0.5176470588235295, 0.3215686274509804),
    (0.3333333333333333, 0.6588235294117647, 0.40784313725490196),
    (0.7686274509803922, 0.3058823529411765, 0.3215686274509804),
    (0.5058823529411764, 0.4470588235294118, 0.7019607843137254),
    (0.5764705882352941, 0.47058823529411764, 0.3764705882352941),
    (0.8549019607843137, 0.5450980392156862, 0.7647058823529411),
    (0.5490196078431373, 0.5490196078431373, 0.5490196078431373),
    (0.8, 0.7254901960784313, 0.4549019607843137),
    (0.39215686274509803, 0.7098039215686275, 0.803921568627451)
] + [new_color]

# Loop through pressures and plot histograms
for j, pressure in enumerate(pressures):
    filename = f"quench340K-{pressure}.txt"
    
    try:
        data = np.genfromtxt(filename).flatten()  # Load and flatten data
        phi = np.concatenate((data, -data))  # Symmetrize the data

        l = int(2 * j / len(pressures))  # Row index
        m = j % int((len(pressures) + 1) / 2)  # Column index

        # Plot histogram
        ax[l, m].text(0, 0.036, '{0:3.1f}'.format(pressure*0.000101325) + " GPa", color=colors[j], ha='center')
        ax[l, m].hist(phi, bins=10, color=colors[j], density=True)
        
        #ax[l, m].yaxis.set_major_formatter(ticker.ScalarFormatter(useMathText=True))
        #ax[l, m].ticklabel_format(style='sci', axis='y', scilimits=(0, 0))  # Force 10^x notation
        #ax[l,m].yaxis.set_offset_position("left")  # Moves the exponent to the top-right
        

        if m == 0:
            ax[l, m].set_ylabel("Probability Density")

        if l == 1:
            ax[l, m].set_xlabel(r"$\langle \phi \rangle$ [degrees]")

    except Exception as e:
        print(f"Error reading {filename}: {e}")

# Remove the last empty subplot if needed
ax[-1, -1].remove()
ax[0, -1].set_xlabel(r"$\langle \phi \rangle$ [degrees]")

plt.ylim([0, 0.035])
plt.savefig("figs/histrograms-340K.pdf", bbox_inches='tight')
plt.show()
```
