```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
font = {'family' : 'serif',
        'serif'   : 'palatino',
        #'sans-serif'    : 'Computer Modern Sans serif',
        'style'   : 'normal',
        'variant'   : 'normal',
        'stretch'   : 'normal',
        'weight'   : 'normal',
        'size'   : 9}
plt.rc('font', **font)
plt.rc('text', usetex=True)
plt.rcParams['figure.figsize'] = (3, 5) # The unit here is inches. 3.5 is good for one column, 7 is good for whole page
```

```python
pressure = np.array([0,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000])

f, ax = plt.subplots(len(pressure),sharex=True,sharey=True)

plt.subplots_adjust(wspace=0.2, hspace=0.3)
new_color = 'seagreen'
#colors = plt.rcParams['axes.prop_cycle'].by_key()['color'] + [new_color]
colors=[(0.2980392156862745, 0.4470588235294118, 0.6901960784313725), (0.8666666666666667, 0.5176470588235295, 0.3215686274509804), (0.3333333333333333, 0.6588235294117647, 0.40784313725490196), (0.7686274509803922, 0.3058823529411765, 0.3215686274509804), (0.5058823529411764, 0.4470588235294118, 0.7019607843137254), (0.5764705882352941, 0.47058823529411764, 0.3764705882352941), (0.8549019607843137, 0.5450980392156862, 0.7647058823529411), (0.5490196078431373, 0.5490196078431373, 0.5490196078431373), (0.8, 0.7254901960784313, 0.4549019607843137), (0.39215686274509803, 0.7098039215686275, 0.803921568627451)] + [new_color]
steps=1000000000

for press,j in zip(pressure,np.arange(len(pressure))):
        filename=str(press) + "atm_1" + "/thermo.txt"
        data = np.genfromtxt(filename)
        ax[j].plot((400 - 300*(data[::1000,0])/steps),np.abs(data[::1000,3]),color=colors[j])
        ax[j].set_ylabel("$|\phi|$")
        ax[j].set_xlim(400,100)
        ax[j].text(90,25,'{0:3.1f}'.format(press*0.000101325) + " GPa",color=colors[j],va='center')
        

ax[-1].set_xlabel("$T$ [K]")

plt.savefig("figs/quenches.pdf",bbox_inches='tight')
plt.show()
```
